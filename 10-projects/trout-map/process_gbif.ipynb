```python
media = pd.read_csv('/Users/clkruse/Downloads/0046033-250827131500795/multimedia.txt', sep='\t')
obs = pd.read_csv('/Users/clkruse/Downloads/0046033-250827131500795/occurrence.txt', sep='\t', on_bad_lines='warn')
print("media columns:")
for col in media.columns:
    print(col)
print("obs columns:")
for col in obs.columns:
    print(col)

print(len(obs), len(media))
```

```python
single_row = obs.iloc[10000]
for col in single_row.index:
    print(f"{col}: {single_row[col]}")
```

```python
simple_out = pd.DataFrame({
    'gbifID': obs['gbifID'],
    'name': obs['vernacularName'],
    'species': obs['species'],
    'date': obs['eventDate'],
    'country': obs['level0Gid'],
    'state': obs['level1Gid'],
    'lat': obs['decimalLatitude'],
    'lon': obs['decimalLongitude'],
    'publisher': obs['publisher']
})

# join the media table to the simple_out table
# The 'gbifID' column in simple_out is the same as the 'gbifID' column in the media table
# I only want to include the 'identifier' column from the media table

simple_media = pd.DataFrame({
    'gbifID': media['gbifID'],
    'format': media['format'],
    'url': media['identifier']
})

# capitalize the first letter of each word in the 'name' column
simple_out['name'] = simple_out['name'].str.title()

merged = pd.merge(simple_out, simple_media, on='gbifID', how='left')

species_to_remove = [
    'Oncorhynchus tshawytscha',
    'Oncorhynchus kisutch',
    'Oncorhynchus nerka',
    'Oncorhynchus gorbuscha',
    'Oncorhynchus keta',
    'Salmo salar'
    ]

# remove species_to_remove
simple_out = simple_out[~simple_out['species'].isin(species_to_remove)]

print(len(simple_out), len(merged))
simple_out['name'].value_counts()
```
